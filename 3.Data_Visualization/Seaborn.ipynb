Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

https://seaborn.pydata.org/

One advantage over matplotlib is that it works natively with DataFrame objects

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

We'll work with predefined data sets

https://en.wikipedia.org/wiki/Iris_flower_data_set


https://github.com/allisonhorst/palmerpenguins

```python
df = sns.load_dataset('iris')
df.head()
```

__Matplotlib vs Seaborn__

```python
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

plt.plot(x, y)
plt.legend('ABCDEF', ncol = 2, loc = 'upper left')
```

```python
z = np.cumsum(rng.randn(500, 6), 0)
df2 = pd.DataFrame(z, columns = ['A', 'B', 'C', 'D', 'E', 'F'])
df2['G'] = np.linspace(0, 10, 500)

sns.lineplot(data=df2,x='G',y='A')
sns.lineplot(data=df2,x='G',y='B')
sns.lineplot(data=df2,x='G',y='C')
sns.lineplot(data=df2,x='G',y='D')
sns.lineplot(data=df2,x='G',y='E')
sns.lineplot(data=df2,x='G',y='F')

plt.legend('ABCDEF', ncol=2, loc='upper left');
```

## Scatter plots

```python
sns.relplot(data = df, x = 'sepal_length', y = 'sepal_width')
```

```python
sns.relplot(data = df, x = 'sepal_length', y = 'sepal_width', style = 'species');
```

```python
sns.relplot(data = df, x = 'sepal_length', y = 'sepal_width', style = 'species', hue = 'species');
```

```python
sns.relplot(data=df,x='sepal_length',y='sepal_width',hue='petal_width');
```

```python
sns.relplot(data=df,x='sepal_length',y='sepal_width',size='petal_width');
```

```python
sns.relplot(data = df, x = 'sepal_length', y = 'sepal_width', kind = 'line');
```

```python
df3 = pd.DataFrame(dict(time = np.arange(500),
                      value = np.random.randn(500).cumsum()))
g = sns.relplot(x = 'time', y = 'value', kind = 'line', data = df3)
```

```python
#no sort
df4 = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df4);
```

# Joint

```python
sns.jointplot(data = df, x = 'sepal_length', y = 'sepal_width',
              kind = 'kde', fill = True);
```
